```python
from functional import curried, compose
import numpy as np
import pandas as pd
import matplotlib
matplotlib.rc("lines", markeredgewidth=0.5) # This is required beacuse: http://bit.ly/2DEqYXt
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

from analysis import *
```

```python
data_train = pd.read_csv('../data/train.csv')
data_test = pd.read_csv('../data/test.csv')
data_all = pd.concat([data_train, data_test], ignore_index=True)
```

```python
df = data_all
df = df.groupby(['Sex', 'Survived'], as_index=False)['PassengerId'].count()
times = df.Sex.unique()
g = sns.FacetGrid(df, col="Survived", palette="Set3", size=4, aspect=2)
g.map(sns.barplot, 'Sex', 'PassengerId', order=times)
```

```python
df = data_all

# Draw a nested barplot to show survival for class and sex
g = sns.factorplot(col="Pclass", x="Survived", hue="Sex", data=df,
                   size=6, kind="count", palette="muted")
g.despine(left=True)
g.set_ylabels("Survived")
```
